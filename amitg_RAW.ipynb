```python
#importing or including important libraries
import pandas as pd
import numpy as np
```

```python
#reading the csv file in a data set
ds = pd.read_csv('raw_data.csv')


#storing the data set in a data frame
df = pd.DataFrame(data=ds)


df
```

```python
df.at[3,'email']='gream@consider.com'
#data.loc[0:9,['Num','NAME']] = [100,'Python']
df.loc[9]=[210,'jim.carry@gmail.com','Male','Amazon','USA',12349,556,23652,67]
```

```python
df
```

```python
df['Country'].replace({'UAS','USA'},inplace=True)
```

```python
df['Country'].fillna(value=df['Country'].mode(),inplace=True)
```

```python
df
```

```python
df['Income.1'] = df['Income.1'].str.replace('$',"")
```

```python
df['Income.1'] = df['Income.1'].replace("",np.nan)
```

```python
df.info()
```

```python
df['Income.1'] = df['Income.1'].astype(float)
```

```python
df['Income.1'].fillna(value=float(df['Income.1'].mean()),inplace=True)
df['Income.2'].fillna(value=float(df['Income.2'].mean()),inplace=True)
```

```python
df['Total Income'] = np.ceil(df['Income.1'] + df['Income.2'])
```

```python
df['DOB'] = pd.to_datetime(df['DOB'], errors='coerce')
df['DOB'] = df['DOB'].dt.strftime('%Y-%m-%d')
```

```python
df.head()
```

```python
df[['First_Name', 'Last_Name']] = df['Employee Full Name'].str.split(' ', expand=True)
df.head()
```

```python
#move column 'd' to the beginning
df = df[['First_Name']+ ['Last_Name'] + [col for col in df.columns if col != 'Last_Name' or col != 'First_Name']]
df.head()
```

```python
result = df.groupby('Total Income')['Company','Country'].aggregate('mean')
print(result)
```
